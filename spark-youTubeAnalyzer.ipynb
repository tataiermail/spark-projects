```python
import os, time
#os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages org.apache.spark:spark-streaming-kafka-0-8_2.11:2.0.2 pyspark-shell'

from pyspark import SparkContext                   #    Spark
from pyspark.sql import *                          #    for SparkSession & Spark SQL
from pyspark.sql.types import *                    #    for StructType
from pyspark.streaming import StreamingContext     #    Spark Streaming
from pyspark.streaming.kafka import KafkaUtils     #    Kafka
import json                                        #    json parsing
```

```python
class youTubeAnalyzer:
    
    def __init__(self):
        self.app_name = "youTubeAnalyzer"
        self.data_file_directory = "C:\\Users\\suman\\Downloads\\Data-Set\\"
        self.yt_file = 'youtubedata.txt'
        
        self.yt_file_schema = StructType([
                                StructField("f1", StringType(), True),
                                StructField("f2", StringType(), True),
                                StructField("f3", DoubleType(), True),
                                StructField("f4", StringType(), True),
                                StructField("f5", DoubleType(), True),
                                StructField("f6", DoubleType(), True),
                                StructField("f7", DoubleType(), True),
                                StructField("f8", DoubleType(), True),
                                StructField("f9", DoubleType(), True),
                                StructField("f10", StringType(), True),
                                StructField("f11", StringType(), True),
                                StructField("f12", StringType(), True),
                                StructField("f13", StringType(), True),
                                StructField("f14", StringType(), True),
                                StructField("f15", StringType(), True),
                                StructField("f16", StringType(), True),
                                StructField("f17", StringType(), True),
                                StructField("f18", StringType(), True),
                                StructField("f19", StringType(), True),
                                StructField("f20", StringType(), True),
                                StructField("f21", StringType(), True),
                                StructField("f22", StringType(), True),
                                StructField("f23", StringType(), True)
                            ])
        
        self.spark = SparkSession.builder.appName("myApp") \
                        .config("spark.mongodb.input.uri", "mongodb://suman1:suman1@ds147480.mlab.com:47480/deliverdb.members") \
                        .config("spark.mongodb.output.uri", "mongodb://suman1:suman1@ds147480.mlab.com:47480/deliverdb.members") \
                        .getOrCreate()
        
    
    def run(self):
        print("================================= ")
        print("Entrypoint to actual application. ")
        
        #Read the YT data from Tab delimited file
        yt_df = self.spark.read.format("csv").option("delimiter", "\t") \
                    .schema(self.yt_file_schema).option("inferSchema", True) \
                    .load(self.data_file_directory + self.yt_file)
    
        yt_df2 = self.spark.read.format("csv").option("delimiter", "\t") \
                    .load(self.data_file_directory + self.yt_file)
    
        return yt_df2
    
```

```python
print("I am a calling function, that will trigger the actual data processing pipeline. ")
yt = youTubeAnalyzer()
print(yt.spark)

yt_df1 = yt.run()
yt_df1.toPandas()

```
